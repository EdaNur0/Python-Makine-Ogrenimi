```python
isimler =['Ben', 'Omer', 'Karen', 'Celine', 'Sue', 'Bora', 'Rose', 'Ellen', 'Bob', 'Taylor,', 'Jude']
arama_sayıları =[300, 10, 500, 70, 100, 100, 600, 800, 200, 450, 80]
anlaşmalar=[7, 10, 15, 6, 7, 6, 8, 7, 13, 6, 7]
gelirler= [2400, 60, 12000, 2275, 500, 770, 4000, 6000, 800, 1200, 500]
```

```python
import numpy as np
```

```python
#dizi oluşturalım
veri = np.array([ ], dtype=int) # içi boş dizi olduğu için [] kullandık
#dtype ile dizide ne tür verilerin olacağını tanımlıyoruz.
```

```python
#veri listesine isim eklemek için oluşruldu
def isim_ekleme(isimler_listesi):
  global veri  #birden fazla yerde veri dizisine değişiklik yapılabilmesi için global kullandık
  for i in isimler_listesi:
    veri = np.append(veri, isimler.index(i))
```

```python
#veri listesine için oluşturuldu
def performans(dokuman_listesi):
  global veri
  veri = np.append(veri, dokuman_listesi) 
```

```python
isim_ekleme(isimler)
performans(arama_sayıları)
performans(anlaşmalar)
performans(gelirler)

print(veri)
print(veri.shape)
```

```python
#Bu yöntem, verilerini değiştirmeden bir diziye yeni bir şekil verir
#reshape()
veri = veri.reshape(4,11)

print(veri)
print(veri.shape)
```

```python
# 1. diziyi yazdır
print(veri[0])
```

```python
#4. diziyi yazdır
print(veri[3])
```

```python
#Ellen'ın ürettiği geliri almak istiyorsak 3. satır ve 7. sütundaki değere erişmemiz gerekiyor.
print(veri[3,7])
```

```python
#Çalışan performansını hesaplamak için bir formülü vardır. Performans= [(Ortalama anlaşma boyutu) * Gelir] / Çağrı sayısı

def çalışan_performans(çalışan_isimleri):
  idx = isimler.index(çalışan_isimleri)
  çağrı_sayı = veri[1, idx]
  ortalama_anlaşma = veri[2,idx]
  gelir = veri[3,idx]

  performans = (ortalama_anlaşma*gelir)/çağrı_sayı
  
  return performans
```

```python
print(çalışan_performans('Ellen'))
```

```python
performans_1 = []
for isim in isimler:
  performans = int(çalışan_performans(isim))
  performans_1.append(performans)
```

```python
#Bu argüman yeni verinin hangi eksen boyunca ekleneceğini belirtmek için kullanılır. 
# 0 değeri dikey ekseni, 1 değeri ise yatay ekseni belirtir.
#.concatenate((..,[...]), axis= 0 or 1)
# "concatenate” yöntemi verilen listeyi dizinin sonuna ekler.
veri = np.concatenate((veri, [performans_1]), axis=0)
print(veri)
```

```python
#En iyi ve en kötü performans gösteren çalışanları bulmak istiyoruz. 
# Numpy'nin .argmin() ve .argmax() yöntemleriyle, belirli bir dizideki minimum ve maksimum değerlerin dizinlerini bulabiliriz.
# .argmin() ---> min       .argmax() ---> max

iyi_durum = np.argmax(veri[4])
kötü_durum = np.argmin(veri[4])

print('En iyi durum: ', iyi_durum)
print('En kötü durum: ', kötü_durum)
```
